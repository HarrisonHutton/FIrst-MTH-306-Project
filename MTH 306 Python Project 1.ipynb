Euler's Method with Python  

```python
# f(x,y) = dy/dx

def f(x,y):
    return y/(x**2)

plt.figure(figsize=(8,8))
slopefieldplot( f, 0.6,2.2, 0.35,0.75, .1, lw=2)

y = 0.4
x = 1
xfinal = 2.5
n = 5
h = (xfinal-x)/n

xlist = [x]
ylist = [y]

for i in range(n):
    slope = f(x,y)
    y = y + h*slope
    x = x + h
    xlist.append(x)
    ylist.append(y)
    
for x,y in zip(xlist, ylist):
    print(f'{x:8.4f} {y:12.8f}')
    
plt.plot( xlist, ylist, 'mo-', lw=3, alpha=0.6, label='Euler approximation')
plt.xlabel('x')
plt.ylabel('y');
```
